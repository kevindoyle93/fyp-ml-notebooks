# Calculating achievable feature alterations

### Sample data of shots on target feature

```python
mean = 4.54
std_dev = 2.63
positive_weight = True
```

### Probability density function (pdf) for shots on target feature

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

from scipy.stats import norm

x = [i for i in range(0, 11)]
dist = norm(mean, std_dev)
plt.plot(x, dist.pdf(x))
plt.title('Probability Density Function: Shots on Target')
plt.xlabel('Number of Shots on Target')
plt.ylabel('pdf(x)')
plt.show()
```

### Using the pdf to make alterations

```python
values = [i for i in range(0, 11)]
alterations = []
for val in values:
    alteration = std_dev ** (dist.pdf(val) * 10)
    alterations.append(alteration)

altered_values = [int(values[i] + alterations[i]) for i in range(0, len(values))]
    
plt.plot(values, alterations, c='r')
plt.plot(values, values, c='b')
plt.plot(values, altered_values, c='g')

plt.title('Achievable Alterations: Shots on Target')
plt.xlabel('Number of Shots on Target')
plt.ylabel('Altered Number of Shots on Target')

green_patch = mpatches.Patch(color='g', label='Altered value')
blue_patch = mpatches.Patch(color='b', label='Initial value')
red_patch = mpatches.Patch(color='r', label='Alteration')
plt.legend(handles=[green_patch, blue_patch, red_patch])

plt.show()
```
