# Density Histograms for Descriptive Features

### Load data from file

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('data/training_data.csv', index_col=0)
df
```

### Function for displaying density histograms of integer count features

```python
def show_integer_density_hist(title, label, data):
    data = data.value_counts(normalize=True)
    
    values = []
    densities = []
    for value, density in sorted(data.items()):
        values.append(value)
        densities.append(density)
    
    y_pos = np.arange(len(densities))
    
    plt.bar(y_pos, densities, align='center', alpha=0.5, color='g')
    plt.ylabel('Density')
    plt.xlabel(label)
    plt.title(title)
    
    plt.savefig('density-histograms/{}.png'.format(title), bbox_inches='tight')
    
    plt.show()
```

### Function for displaying density histograms of percentage features

```python
def show_percentage_density_hist(title, label, data):
    plt.hist(data, alpha=0.5, color='g', bins=20, edgecolor='w')
    plt.title(title)
    plt.xlabel(label)
    plt.ylabel('Density')
    
    plt.savefig('density-histograms/{}.png'.format(title), bbox_inches='tight')
    
    plt.show()
```

### Half Time Goals

```python
show_integer_density_hist('Home Half Time Goals', 'Number of goals', df['half_time_home_goals'])
show_integer_density_hist('Away Half Time Goals', 'Number of goals', df['half_time_away_goals'])
```

### Possession

```python
show_percentage_density_hist('Home Possession', 'Percentage possession', df['home_possession'])
show_percentage_density_hist('Away Possession', 'Percentage possession', df['away_possession'])
```

### Total Shots

```python
show_integer_density_hist('Home Total Shots', 'Number of shots', df['home_total_shots'])
show_integer_density_hist('Away Total Shots', 'Number of shots', df['away_total_shots'])
```

### Corners

```python
show_integer_density_hist('Home Corners', 'Number of corners', df['home_corners'])
show_integer_density_hist('Away Corners', 'Number of corners', df['away_corners'])
```

```python
show_integer_density_hist('Home Fouls', 'Number of Fouls', df['home_fouls_committed'])
show_integer_density_hist('Away Fouls', 'Number of Fouls', df['away_fouls_committed'])
```
